**4.** $\beta = 5^{\frac13} - 2 \times 5^\frac13$ est un élément de $\mathbb{Q}(5^\frac13)$. Maintenant le polynôme minimal de $5^\frac13$ est $X^3 - 5 \in \mathbb{Q}[X]$. Ensuite $\beta = B(5^\frac13)$, avec $B(X) = X - 2X^2 \in \mathbb{Q}[X]$. On applique donc les questions précédentes :

```python
A.<X, Y> = PolynomialRing(QQ, order = 'lex')
I = A.ideal(Y - X + 2*X^2, X^3 - 5)
J = I.elimination_ideal(X)
Pbeta = J.gens()[0]
Pbeta
```
