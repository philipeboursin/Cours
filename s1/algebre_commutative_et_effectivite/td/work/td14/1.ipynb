**Exercice 1**

Rappelons que si $G$ est une base de Groebner marquée $\{(g_i, \alpha_i)\}$, où $g_i = x^{\alpha_i} + \sum_{\beta} c_{i, \beta}x^\beta$, alors l'adhérence du cone ouvert associé dans l'éventail de Groebner est donné par $\overline{C}_G = \{w \in \mathbb{R}^n_+ \mid \alpha_i \cdot w \geq \beta \cdot w,\, \forall \beta$ tq $c_{i, \beta} \neq 0\}$.

**1)**

```python
A.<x,y> = PolynomialRing(QQ, order = 'lex')
I = A.ideal(x^2 + y^2 - 1, x + 2*y)
```

```python
G1 = I.groebner_basis()
G1
```

Ainsi $w = (w_1, w_2) \in \overline{C}_{G_1}$ si et seulement si $w \in \mathbb{R}^2_+$ et 
\begin{align*}
    \begin{cases}
    w_1 \geq w_2 \\
    w_2 \geq 0 \\
    \end{cases}
\end{align*}

```python
A = A.change_ring(names = 'y, x')
I = I.change_ring(A)
G2 = I.groebner_basis()
G2
```

Ainsi $(w_1, w_2) \in \overline{C}_{G_2}$ si et seulement si
\begin{align*}
    \begin{cases}
        w_2 \geq w_2 \\
        w_1 \geq 0 \\
    \end{cases}
\end{align*}

On a donc trouvé tous les cônes de l'éventail de Groebner, et graphiquement cela donne

**2)** Calculons trois bases de Groebner pour trois ordres différents :

```python
A.<x,y> = PolynomialRing(QQ, order = 'lex')
I = A.ideal(x^3 - y, x + y^3 + 1)
G1 = I.groebner_basis()
print(G1)
B = A.change_ring(names = 'y,x')
G2 = I.change_ring(B).groebner_basis()
print(G2)
B = A.change_ring(names = 'x,y', order = 'degrevlex')
G3 = I.change_ring(B).groebner_basis()
print(G3)
```
