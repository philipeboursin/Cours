```python
A.<x, y> = PolynomialRing(QQ, order = 'lex')
f = x^4 + y^4 - 1
g = x^5*y^2 - 4*x^3*y^3 + x^2*y^5 - 1
```

```python
res = f.resultant(g, y)
R = res.univariate_polynomial().roots(RR, multiplicities = false)
S = []
r = R[0]

B.<x> = RR[]


for r in R:
    R1 = B(f(y = r)).roots(multiplicities = false)
    R2 = B(g(y = r)).roots(multiplicities = false)
    inter = list(set(R1) & set(R2))
    for i in inter:
        S.append((i, r))

print(S)
```

```python
p = implicit_plot(f, (x, -2, 2), (y, -2, 2))  + implicit_plot(g, (x, -2, 2), (y, -2, 2))
for s in S:
    p = p + point(s, color = 'red', size = 40)
    
p
```
